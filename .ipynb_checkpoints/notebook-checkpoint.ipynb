# Modelo para salida de clientes en Beta bank

## Instroducción

Los clientes de Beta Bank se están yendo, cada mes, poco a poco. Los banqueros descubrieron que es más barato salvar a los clientes existentes que atraer nuevos. Necesitamos predecir si un cliente dejará el banco pronto.
Crearemos un modelo con el máximo valor **F1** posible (al menos **0.59**).

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.utils import shuffle
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.metrics import classification_report, f1_score, precision_score, recall_score
from sklearn.metrics import confusion_matrix, roc_auc_score, roc_curve

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv('/datasets/Churn.csv')
```

```python
# Estableciendo el random_state para reproducibilidad.
random_state = 262626
```

**Características**

*   _RowNumber_: índice de cadena de datos
*   _CustomerId:_ identificador de cliente único
*   _Surname:_ apellido
*   _CreditScore:_ valor de crédito
*   _Geography:_ país de residencia
*   _Gender:_ sexo
*   _Age:_ edad
*   _Tenure:_ período durante el cual ha madurado el depósito a plazo fijo de un cliente (años)
*   _Balance:_ saldo de la cuenta
*   _NumOfProducts:_ número de productos bancarios utilizados por el cliente
*   _HasCrCard:_ el cliente tiene una tarjeta de crédito (1 - sí; 0 - no)
*   _IsActiveMember:_ actividad del cliente (1 - sí; 0 - no)
*   _EstimatedSalary:_ salario estimado

**Objetivo**

*   _Exited:_ El cliente se ha ido (1 - sí; 0 - no)

```python
data.info()
data.head()
```

**Observaciones**

Tenemos 14 caolumnas con diez mil observaciones, identificamos a *Exited* como nuestra variable objetivo, podemos observar también valores ausentes en la variable *Tenure*.

## Preparación de datos

**Descartando caracteristicas no necesarias**

Vamos a descartar tres caracteristicas de nuestro conjunto de datos que no nos seran de utilidad. 

```python
data.drop(columns=['RowNumber', 'CustomerId', 'Surname'], inplace=True)
```

```python
data.head()
```

Nos quedaremos con 10 caracteristicas y nuestra variable objetivo.

**División de caracteristicas y objetivo**

Hacemos la separación de nuestras caracteristicas y nuestra variable objetivo.

```python
features = data.drop(columns='Exited', axis=0)
target = data['Exited']
```

**Valores ausentes**

Notamos la presentacia de valores ausentes en la variable *Tenure*, realizaremos una imputación de estos valores por la media, y para mantener infomración sobre estos valores ausentes crearemos una nueva variable que guarde '1' en las observaciones que inicialmente tenían un valor ausente.

```python
features.isna().sum()
```

```python
features['Tenure_is_na'] = features['Tenure'].isna().astype(int)
```

```python
tenure_mean = features['Tenure'].mean()
features['Tenure'] = features['Tenure'].fillna(tenure_mean)
```

**Codificación y estandartización**

Vamos a procesar nuestras caracteristicas para dejarlas listas para proporcinarlas a los diferentes modelos, pese a estar estos procesos enfocados en modelos de regresión logística, ya que estos no pueden procesar caracteristicas categoricas terminaremos utilizando otro tipos de modelos como *Random Forest*.

Consideraremos tres procedimientos:
1. Para nuestras caracteristicas categóricas emplearemos One-Hot encoding.
2. Para las caracteristicas numericas realizaremos el escalado de caracteristicas.
3. En las caracteristicas binarias, las pasaremos sin modificación para mantener esta binariedad.

```python
cat_features = features.columns[features.dtypes == object].to_list()
num_features = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']
bool_features = ['HasCrCard','IsActiveMember', 'Tenure_is_na']
```

```python
ct = ColumnTransformer(transformers=[
        ('numeric', StandardScaler(), num_features),
        ('categoical', OneHotEncoder(drop='first'), cat_features),
        ('boolean', 'passthrough', bool_features)
    ])
```

```python
transformed_features = ct.fit_transform(features)
columns_names = num_features + ct.transformers_[1][1].get_feature_names_out(cat_features).tolist() + bool_features

processed_features = pd.DataFrame(transformed_features, columns=columns_names)
```

```python
processed_features.head()
```

Tenemos ahora neustras caracteristicas procesadas, debido a aplicar *OHE* ahora tenemos 12 caracteristicas.

**Separación de datos**

Realizaremos la separación de nuestro conjunto en uno de entrenamiento, validación y prueba.

```python
features_train, features_valid_test, target_train, target_valid_test = train_test_split(processed_features, target,
                                                                                        test_size=0.4,
                                                                                        random_state=random_state,
                                                                                        stratify=target)
features_valid, features_test, target_valid, target_test = train_test_split(features_valid_test, target_valid_test,
                                                                                        test_size=0.5,
                                                                                        random_state=random_state,
                                                                                        stratify=target_valid_test)
```

```python
print(f'df train: {features_train.shape[0]/processed_features.shape[0]:0.1%}')
print(f'df valid: {features_valid.shape[0]/processed_features.shape[0]:0.1%}')
print(f'df test:  {features_test.shape[0]/processed_features.shape[0]:0.1%}')
```

### Analizando desbalance de clase


```python
def upsample(_features, _target, repeat):
    """
    Realiza un sobremuestreo de las observaciones positivas (1).
    
    Args:
        _features (DataFrame): Un DataFrame que contiene las características.
        _target (Series): Una Serie que contiene las etiquetas objetivo (0 o 1).
        repeat (int): El número de veces que se deben repetir la clase minoritaria (1).

    Returns:
        features_upsampled (DataFrame): Un DataFrame con características remuestreadas.
        target_upsampled (Series): Una Serie con etiquetas de destino remuestreadas.

   """
    features_zeros = _features[_target == 0]
    features_ones = _features[_target == 1]
    target_zeros = _target[_target == 0]
    target_ones = _target[_target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)

    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=random_state)

    return features_upsampled, target_upsampled

def downsample(_features, _target, fraction):
    """
    Realiza un submuestreo de las observaciones negativas (0).

    Args:
        _features (DataFrame): Un DataFrame que contiene las características de los datos.
        _target (Series): Una Serie que contiene las etiquetas de destino (0 o 1).
        fraction (float): La fracción de muestras de la clase mayoritaria (0) para el submuestreo.

    Returns:
        features_downsampled (DataFrame): Un DataFrame con características remuestreadas.
        target_downsampled (Series): Una Serie con etiquetas de destino remuestreadas.
    """
    features_zeros = _features[_target == 0]
    features_ones = _features[_target == 1]
    target_zeros = _target[_target == 0]
    target_ones = _target[_target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=random_state)]
        + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=random_state)]
        + [target_ones])

    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=random_state)

    return features_downsampled, target_downsampled

def stats_model(_model, _features_train, _target_train, _features_valid=features_valid, _target_valid=target_valid):
    """
    Esta función evalúa un modelo de aprendizaje automático y muestra varias métricas de rendimiento.
    Grafica la matriz de confusión y la Curva ROC, además de mostrar el valor de F1 Score y AUC-ROC.

    Args:
        _model: El modelo de aprendizaje automático que se va a evaluar.
        _features_train (DataFrame): Un DataFrame que contiene las características de entrenamiento.
        _target_train (Series): Una Serie que contiene las etiquetas de destino de entrenamiento.
    """
    model = _model
    model.fit(_features_train, _target_train)
    predicted_valid = model.predict(_features_valid)
    
    print(classification_report(_target_valid, predicted_valid))
    print('F1 Score:',f1_score(_target_valid, predicted_valid))
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,6))
    sns.heatmap(confusion_matrix(_target_valid, predicted_valid), annot=True, fmt='g', cmap='Blues', ax=ax1)
    ax1.set_ylabel('predictions')
    ax1.set_xlabel('real values')
    ax1.set_title("Matriz de confusión")
    
    probabilities_valid = _model.predict_proba(_features_valid)
    probabilities_one_valid = probabilities_valid[:, 1]

    auc_roc = roc_auc_score(_target_valid, probabilities_one_valid)
    print('AUC-ROC:',auc_roc)

    fpr, tpr, thresholds = roc_curve(_target_valid, probabilities_one_valid)

    # Curva de modelo
    ax2.plot(fpr, tpr)
    # Curva ROC para modelo aleatorio
    ax2.plot([0, 1], [0, 1], linestyle='--')
    ax2.set_xlim([0.0, 1.0])
    ax2.set_ylim([0.0, 1.0])
    ax2.set_xlabel("Tasa de falsos positivos")
    ax2.set_ylabel("Tasa de verdaderos positivos")
    ax2.set_title("Curva ROC")

def show_proportion_target(_target):
    """
    Esta función muestra la proporción de las etiquetas de destino en un conjunto de datos, imprime una
    gráfica de pastel con las proporciones.

    Args:
        _target (Series): Una Serie que contiene las etiquetas de destino (0 o 1).
    """
    print(_target.value_counts())
    _target.value_counts(normalize=True).plot(kind='pie', autopct='%.0f%%')
    plt.show()

def test_rf_model(_features_train, _target_train):
    """
    Esta función busca el mejor modelo RandomForest evaluando diferentes combinaciones de hiperparámetros.

    Args:
        _features_train (DataFrame): Un DataFrame que contiene las características de entrenamiento.
        _target_train (Series): Una Serie que contiene las etiquetas de destino de entrenamiento.
    
    Returns:
        result (str): Un mensaje que indica el mejor modelo encontrado junto con sus hiperparámetros.
        best_f1 (float): El valor de la mejor puntuación F1 encontrada.
        best_est (int): El número de estimadores del mejor modelo.
        best_depth (int): La profundidad máxima del mejor modelo.
    """
    best_f1 = 0
    best_depth = 0
    best_est = 0
    results_rf = pd.DataFrame()
    for est in [10, 100, 200]:
        for depth in range(5, 26, 5):
            model = RandomForestClassifier(random_state=random_state, n_estimators=est, max_depth=depth)
            model.fit(_features_train, _target_train)
            predicted_valid = model.predict(features_valid)
            f1 = f1_score(target_valid, predicted_valid)
            results_rf.loc[depth, f'est-{est}'] = f1
            if f1 > best_f1:
                best_f1 = f1
                best_depth = depth
                best_est = est

    result = f'Mejor modelo F1={best_f1:0.2f} | [estimators: {best_est}, max_depth: {best_depth}]'
    return result, best_f1, best_est, best_depth
    
```

Observamos el balance de nuestra variable objetivo.

```python
show_proportion_target(target_train)
```

Tenemos una clase mayoritaria que nos deja un *desbalance de clase*, esto puee influir en nuestra métrica de *F1*, llevemos a cabo un modelo de *Regresión Logistica* que no haga distinción de esto y veamos que resultados obtenemos.

```python
stats_model(LogisticRegression(random_state=random_state,
                               solver='liblinear'), 
            features_train, target_train)
```

**Observaciones**

Vemos que al implementar nuestro mondelo sin considerar el desbalance de clase, tenemos una *exactitud* de *0.81*, sin embargo en **F1** tenemos apenas **0.31**, esto puede deberse a que acierta bastante con la clase que predomina pero tiende a fallar más con la que tiene menos observaciones.

Tal como anticipamos, observando la matriz de confusión es más evidente como acierta en gran proporción las predicciones en 0 respecto en aquellas en que falla, sin embargo para las predicciones para 1, termina fallando bastante más de las que acierta.

En nuestra métrica *AUC-ROC* tenemos un valor de *0.75* y al visualizar la curva vemos que nuestro modelo está por encima de un clasificador aleatorio al menos para identificar cada clase, sin embargo la puntuación baja en *F1* que es especialmente sensible a los VP y FP-FN, nos indica que tiende a errar bastante cuando se trata de predecir los casos positivos.

### Balanceando clases

Vamos a probar distintos métodos y modelos para encontrar el que nos pueda ofrecer una puntuación de *F1* superior.

**LogisticRegression: class_weight**

Como punto de partida observemos el modelo si indicamos que la propiedad *class_weight='balanced'* para que haga un ajuste automático del peso de las clases.

```python
stats_model(LogisticRegression(class_weight='balanced',
                               random_state=random_state,
                               solver='liblinear'),
            features_train, target_train)
```

**Observaciones**

Nuestro modelo de regresión logística empleando el argumento *class_weight*, en relación a aquel que no hacía distinción del desbalance en las clases mejora en **F1** con **0.49**.  

**LogisticRegression: Sobremuestreo**

Siguiendo otro método para mitigar el desbalance ahora vamos a realizar un remuestreo, en este caso multiplicando las observaciones con nuestra clase 1 para que estas tengan mas observaciones. Vamos a duplicar las observaciones en la clase minoritaria (1), en vez de solo equiparar las observaciones en la clase 0, con el fin de evitar tener algún *sobreajuste*.

```python
features_upsampled, target_upsampled = upsample(features_train, target_train, 2)
```

```python
show_proportion_target(target_upsampled)
```

```python
stats_model(LogisticRegression(random_state=random_state,
                               solver='liblinear'),
            features_upsampled, target_upsampled)
```

**Observaciones**

Comenzamos a notar mejoras en nuestra métrica **F1**, ahora aplicando un sobremuestreo la puntuaciíon subió hasta **0.47**. 

**LogisticRegression: submuestreo**

Siguiendo con el remuestreo ahora haremos un submuestreo, bajando la cantidad de observaciones en la clase 0 para tener una proporción similar entre ambas.

```python
features_downsampled, target_downsampled = downsample(features_train, target_train, 0.25)
```

```python
show_proportion_target(target_downsampled)
```

```python
stats_model(LogisticRegression(random_state=random_state, 
                               solver='liblinear'),
            features_downsampled, target_downsampled)
```

**Observaciones**

Obtenemos nuevamente una mejor puntuación de **F1**, ahora con **0.47**, Es importante mencionar como nuestra métrica de *AUC-ROC* se ha visto consistente entre estos modelos, presentando apenas cambios, lo que nos puede indicar que nuestro modelo no ha perdido la capacidad de identificar cada clase, pero la mejora en *F1* indica que vamos reduciendo los errores y mejorando los *Verdaderos Positivos*.

**RandomForestClassifier: buscando mejores hiperparametros**

Vamos a probar con modelos de *Random Forest* para ver si podemos encontrar un modelo con mejor *F1*, llevaremos a cabo una prueba validando diferentes modelos para encontrar la combinación de hiperparametros que nos de la mejor puntuación para *F1*.

```python
%%time
result, best_f1, best_est, best_depth = test_rf_model(features_train, target_train)
result
```

```python
stats_model(RandomForestClassifier(random_state=random_state, 
                                   n_estimators=best_est, 
                                   max_depth=best_depth), 
            features_train, target_train)
```

**Observaciones**

Obtenemos un incremento considerable en la métrica **F1** con **0.57** en nuestro modelo de *Random Forest*, también se ve incrmentada *AUC-ROC*. 

**RandomForestClassifier: sobremuestreo**

Ahora busquemos un modelo de *Random Forest* entrenando el modelo con nuestros datos de entrenamiento al que hicimos un sobremuestreo.

```python
%%time
result, best_f1, best_est, best_depth = test_rf_model(features_upsampled, target_upsampled)
result
```

```python
stats_model(RandomForestClassifier(random_state=random_state,
                                   n_estimators=best_est,
                                   max_depth=best_depth),
            features_upsampled, target_upsampled)
```

**Observaciones**

Obtuvimos un valor de **F1** mejor, de **0.60**, debido al sobremuestreo y la cantidad de profundidad pudiera tener problema de sobreajuste, aunque *Random Forest* tiene la capacidad de reducir esto, también tratamos nuestros datos con el sobremuestreo de manera mesurada para no incrementar las probabildiades de esto.

**RandomForestClassifier: submuestreo**

Ahora hagamos una prueba con nuestro modelo de *Random Forest* entretandolo con nuestro conjunto al que le hicimos un submuestreo.

```python
%%time
result, best_f1, best_est, best_depth = test_rf_model(features_downsampled, target_downsampled)
result
```

```python
stats_model(RandomForestClassifier(random_state=random_state,
                                   n_estimators=best_est,
                                   max_depth=best_depth),
            features_downsampled, target_downsampled)
```

**Observaciones**

Nuestro modelo de *Random Forest* entrenado con los datos de sobremuestreo solo nos da una medida de **F1** de **0.58**, inferir al modelo anterior.

### Probando modelo final

Tras probar diferentes modelos, el que mejor puntuación de **F1** nos dió fue *Random Forest* utilizando nuestros datos que aplicamos un *sobremuestreo* con los hiperparametros El modelo con la mejor *exactitud* fue con nuestro modelo de **RandomForest**, con los hiperparámetros; **n_estimators = 200** y  **max_depth = 10**.

Realizemos la evaluación final de este, para el entrenamiento final uniremos los datos de entrenamiento con los de validación y realizaremos las pruebas con el conjunto de datos de prueba. 

```python
final_features_train = pd.concat([features_upsampled, features_valid])
final_target_train = pd.concat([target_upsampled, target_valid])
```

```python
stats_model(RandomForestClassifier(random_state=random_state,
                                   n_estimators=200,
                                   max_depth=10),
            final_features_train, final_target_train,
            features_test, target_test)
```

El valor de **F1** de nuestro modelo final es de **0.63**, y un valor de **AUC-ROC** de **0.87**.

### Conclusión

El modelo con el que comenzamos que tenía una buena métrica de **AUC-ROC** pero un valor muy bajo de **F1** mejoró tras hacer procedimientos para mitigar el desbalance e clase, además de probar difernetes modelos para encontrar el mejor. La métrica final de **F1** para nuestro modelo fue de **0.63**, casi el doble del modelo que no tenía ningún tratamiento para el desbalance de clase, que fue de apenas *0.31*. Nuestro modelo mejoró en clasificar los valores positivos, y el valor de **AUC-ROC** fue de **0.87**, el cual también mejoró respecto al primer modelo nos puede ayudar a saber que tan bueno es para disnguir entre ambas clases, que fue mayormente bueno en todo momento, y como es observable en la gráfica estando bastante por encima de un modelo aleatorio.
